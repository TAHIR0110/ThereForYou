```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

##  Univariate Analysis
Univariate analysis refers to the statistical examination of a single variable. The term "univariate" comes from combining "uni," meaning one, and "variate," meaning variable.

```python
# Plotting Age at Diagnosis
plt.figure(figsize=(10, 6))
sns.histplot(data['Diagnosis Age'], bins=30, kde=True)
plt.title('Distribution of Age at Diagnosis')
plt.xlabel('Age at Diagnosis')
plt.ylabel('Frequency')
plt.show()
```

## Univariate Analysis: Age at Diagnosis
- The distribution of age at diagnosis is approximately normal but slightly right-skewed.
- The majority of patients are diagnosed between the ages of 50 and 70.


```python
# Plotting Mutation Count
plt.figure(figsize=(10, 6))
sns.histplot(data['Mutation Count'], bins=50, kde=True)
plt.title('Distribution of Mutation Count')
plt.xlabel('Mutation Count')
plt.ylabel('Frequency')
plt.show()
```

## Univariate Analysis: Mutation Count
- The mutation count distribution is heavily right-skewed, with a few patients having very high mutation counts.
- Most patients have mutation counts below 2,500.


## Multivariate Analysis
Multivariate analysis is a set of statistical techniques used to analyze data that involves multiple variables simultaneously. It allows researchers to understand relationships between variables, patterns within the data, and the structure of complex datasets.

```python
# Correlation Matrix
# Select only numeric columns for the correlation matrix
numeric_data = data.select_dtypes(include=[np.number])

# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Plotting the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

## Multivariate Analysis: Correlation Matrix
- There is a strong positive correlation between MSI MANTIS Score and Mutation Count.
- The Fraction Genome Altered shows a moderate positive correlation with Mutation Count.
- Diagnosis Age shows little to no correlation with Mutation Count or MSI scores.


## Checking Distribution (Skewness)
Checking the distribution of a dataset, particularly skewness, involves assessing how much the data deviates from a symmetrical, bell-shaped (normal) distribution. Skewness is a measure of the asymmetry of the distribution of values.

```python
# Skewness
print(data[['Diagnosis Age', 'Mutation Count', 'Fraction Genome Altered', 'MSI MANTIS Score', 'MSIsensor Score']].skew())

# Plotting the distributions
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
sns.histplot(data['Diagnosis Age'], bins=30, kde=True, ax=axes[0, 0])
sns.histplot(data['Mutation Count'], bins=30, kde=True, ax=axes[0, 1])
sns.histplot(data['Fraction Genome Altered'], bins=30, kde=True, ax=axes[0, 2])
sns.histplot(data['MSI MANTIS Score'], bins=30, kde=True, ax=axes[1, 0])
sns.histplot(data['MSIsensor Score'], bins=30, kde=True, ax=axes[1, 1])
plt.tight_layout()
plt.show()
```

## Skewness of Columns
- **Diagnosis Age**: Slightly right-skewed.
- **Mutation Count**: Heavily right-skewed.
- **Fraction Genome Altered**: Right-skewed.
- **MSI MANTIS Score**: Right-skewed.
- **MSIsensor Score**: Right-skewed.


## Detecting Outliers
Detecting outliers involves identifying observations in a dataset that significantly deviate from the rest of the data. Outliers can occur due to various reasons such as data entry errors, measurement errors, or genuine anomalies in the data.

```python
# Boxplots to detect outliers
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
sns.boxplot(data['Diagnosis Age'], ax=axes[0, 0])
sns.boxplot(data['Mutation Count'], ax=axes[0, 1])
sns.boxplot(data['Fraction Genome Altered'], ax=axes[0, 2])
sns.boxplot(data['MSI MANTIS Score'], ax=axes[1, 0])
sns.boxplot(data['MSIsensor Score'], ax=axes[1, 1])
plt.tight_layout()
plt.show()
```

## Outliers Detection
- **Diagnosis Age**: Few outliers at the higher age spectrum.
- **Mutation Count**: Significant number of outliers with high mutation counts.
- **Fraction Genome Altered**: Presence of outliers with higher altered fractions.
- **MSI MANTIS Score**: Outliers present with higher MSI scores.
- **MSIsensor Score**: Similar to MSI MANTIS, outliers with high scores are observed.

Outliers indicate extreme values in mutation counts and MSI scores, which may need special attention during analysis.


## Class Label Imbalance
Class label imbalance, also known as class imbalance, occurs when one class (or label) in a classification problem has significantly fewer observations than the other classes. This imbalance can skew the predictive model's performance and bias it towards the majority class.

```python
# Plotting Overall Survival Status
plt.figure(figsize=(6, 4))
sns.countplot(x='Overall Survival Status', data=data)
plt.title('Overall Survival Status Distribution')
plt.xlabel('Overall Survival Status')
plt.ylabel('Count')
plt.show()
```

## Class Label Imbalance: Overall Survival Status
- There is a class imbalance in the overall survival status with a higher number of living patients (84%) compared to deceased (16%).
